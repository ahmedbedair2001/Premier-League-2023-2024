```python
epl.drop(['Unnamed: 0','Notes'],axis=1,inplace=True)
```

```python
epl['schedule']=''
```

```python
for i in range (epl.shape[0]):
    if(epl['Day'][i]=='Sat' or epl['Day'][i]=='Sun')and(epl['Time'][i]<'15:00'):
         epl['schedule'][i]='Early'
    elif(epl['Day'][i]=='Sat')and(epl['Time'][i]<'17:30'):
         epl['schedule'][i]='regular'
    elif(epl['Day'][i]=='Sat')and(epl['Time'][i]<'20:00'):
         epl['schedule'][i]='evening'
    elif(epl['Day'][i]=='Sat')and(epl['Time'][i]>='20:00'):
         epl['schedule'][i]='Late'
    elif(epl['Day'][i]=='Sun')and(epl['Time'][i]>='15:00'):
         epl['schedule'][i]='Late'
    elif(epl['Day'][i]=='Tue') and(epl['Day'][i]=='Wed'):
        epl['schedule'][i]='Midweek'
    else:
        epl['schedule'][i]='Occasional'
```

```python
epl.to_csv('epl matches updated.csv',index=False)
```
